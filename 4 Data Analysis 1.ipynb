# NumPy

NumPy (Numerical Python) is the foundational library for scientific computing with Python. NumPy introduces a new data structure, the `ndarray` that provides efficient computation due their homogeneity. This creates for some limitations compared to Python lists but the improvements in speed and useability outweigh the drawbacks.

# Pandas

Pandas is the defacto data analysis library for Python. There are many similarities between Excel and Pandas that we will explore.  
However, Pandas is not an Excel replacement. They both have their place and complement each other nicely.  
We will see that some things are still easier to see with Excel, but with Pandas you get to use the entire Python ecosystem which can be very powerful.

https://pandas.pydata.org/pandas-docs/stable/

https://pandas.pydata.org/pandas-docs/stable/getting_started/10min.html

Pandas provides two new data structures: Series and DataFrame. A dataframe is similar to a spreadsheet, as in a matrix of data with rows and columns. A series can be thought of as a single __column__ (technically an array) of data which is typically the same data type (but doesn't have to be). Series are very similar to numpy arrays and we will see that it is simple to switch between them.

```python
import pandas as pd
```

## Series

```python
s = pd.Series([100, 200, 300])
```

```python
s
```

Notice the index starts at 0, similar to a Python list.  
And since the series is all integers, pandas has made the data type `dtype` of integers `int64`.

This has perfomance improvements over a typical Python list.

```python
s.values # return a numpy array
```

```python
s[0]
```

```python
s[:2]
```

The big difference between Python lists and series is that series perform element-wise calculations:

```python
l1 = [100, 200, 300] 
l2 = [400, 500, 600];
l1 + l2
```

```python
s1 = pd.Series(l1)
s2 = pd.Series(l2)
s1 + s2
```

```python
#We can still do concatnation like lists, the syntax is just different:
pd.concat([s1, s2], ignore_index=True)
```

# Missing data
The 2 month term values are missing for May 2018 rates. Let's fill the missing values using linear interpolation between the 1 month and 3 month tenors.  
More methods can be found here: https://pandas.pydata.org/pandas-docs/stable/user_guide/missing_data.html

```python
df_yc_201805.interpolate(method='linear', axis=1, inplace= True); df_yc_201805
```

axis=1 means to operate on columns in each row, 0 means row-level operations

Most pandas functions return a new dataframe and leave the original one unchanged,  
but with the inplace flag the behaviour is to change the original dataframe.


```python
df_yc_201805.T.plot(figsize=(20,10)) # note the chaining of the Transpose and plot command.
```
